**1. What is a parameter?**
 - A parameter is a configuration variable that is internal to the model and whose value is estimated from data during training. In Machine Learning, parameters are values like weights in linear regression or coefficients in logistic regression that the model learns.

**2. What is correlation?**
- Correlation is a statistical measure that describes the strength and direction of a relationship between two variables. A high correlation means the variables move together, while a low correlation means they move independently.

**3. What does negative correlation mean?**
 - Negative correlation means that as one variable increases, the other variable decreases. For example, the number of hours spent studying and the number of errors on a test could have a negative correlation.

**17. What is causation? Explain difference between correlation and causation with an example.**

 - Causation means one variable directly causes the change in another variable.

 - Correlation: Two variables are related but one may not cause the other.

 - Causation: One variable directly affects the other.

**Example:***

 - Correlation: Ice cream sales and drowning rates both rise in summer (related, but not caused).

 - Causation: Smoking causes lung cancer (direct cause-effect).

**18. What is an Optimizer? What are different types of optimizers? Explain each with an example.**

 - An optimizer adjusts the model's parameters to minimize the loss function.

**Types of Optimizers:**

 - Gradient Descent: Updates parameters by calculating gradients.

 - Stochastic Gradient Descent (SGD): Updates parameters using one sample at a time.

 - Adam Optimizer: Combines momentum and RMSProp; adapts learning rate.
